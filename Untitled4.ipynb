```python
#descrição = describe
jogadores = pd.read_csv("/content/jogadores - jogadores.csv")
jogadores = pd.DataFrame(jogadores)
display(jogadores.describe())
```

```python
#Qual a média da altura dos jogadores
jogadores["Altura em cm"].mean()
```

```python
#Qual a maior altura dos jogadores
jogadores["Altura em cm"].max()
```

```python
#Qual a menor altura dos jogadores
jogadores["Altura em cm"].min()
```

```python
#Qual o jogador mais leve para o mais pesado
jogadores["IMC"] = jogadores["Peso em kg"]/jogadores["Altura em cm"] **2
display(jogadores)
```
