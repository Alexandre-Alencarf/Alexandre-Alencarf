```python
import pandas as pd
pd.options.display.float_format = "{:,.2f}".format
```

```python
linguagens_df = pd.DataFrame({
    "Linguagem": ["Python", "R", "SQL", "C", "Java", "Javascript", "Elixir"], 
    "Popularidade": ["Alta", "Baixa", "Altíssima", "Média", "Altíssima", "Altíssima", "Baixa"],
    "Versão" : [3.10, 4.2, None, None, 11, 16, 1.12],
    "Desenvolvedores" : [7_000_000, 500_000, 20_000_000, 3_500_000, 16_000_000, 25_000_000, 200_000],
    "Data de lançamento" : ["1991-02-20", "1993-08-03", "1986-01-01", "1972-01-01", "1995-05-23", "1995-12-04", "2012-01-01"]
})
```

```python
dfs = pd.DataFrame(
    data  = [],
    columns = ["Lógica de Programação", "Python", "Data Science", "HTML e CSS", "JavaScript", "React Native"],
    index=[list(range(1,11))]
)
dfs
```

```python
dfs["Alunos"] = dfs.mean(axis="columns")
dfs
```

```python
dfs.drop(["Alunos"], axis = "columns", inplace=True)
```

```python
dfs
```

```python
import numpy as np
dfs = dfs.apply(lambda nota: np.random.uniform(low=5.0, high=10.0, size=10), axis="index")
dfs
```

```python
dfs.insert(0, "Nome Aluno", ["Maria", "Paulo", "João", "Ana", "Lívia", "Fernanda", "Telmo", "Otávio", "Lúcia", "Maria"], True)
dfs
```

```python
dfs.columns.values[0] = "Aluno"
dfs
```

```python
dfs["Média"] = dfs.mean(axis = "columns")
dfs
```

```python
dfs["Situação"] = dfs["Média"].map(lambda media: "AP" if media>7 else "RP")
dfs
```

```python
dfs.rename(str.upper, axis="columns", inplace=True)
dfs
```

```python
linguagens_df.rename(str.lower, axis="columns", inplace=True)
linguagens_df
```

```python
linguagens_df.rename(columns={"linguagem":"linguagens"}, inplace=True)
linguagens_df
```

```python
linguagens_df.columns=linguagens_df.columns.str.replace(" ", "_")
linguagens_df
```

```python
dfs.columns=dfs.columns.str.lower().str.replace(" ", "_")
dfs
```

```python
linguagens_df.rename(
    index={0:"A", 1:"B", 2:"C", 3:"D", 4:"E", 5:"F", 6:"G"},
    inplace=True
)
linguagens_df
```

```python
linguagens_df["linguagens"]=linguagens_df["linguagens"].apply(lambda txt: txt.upper())
linguagens_df
```
