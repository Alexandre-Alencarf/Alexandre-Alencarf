```python
import pandas as pd
```

```python
dataset = r"/content/base_funcionarios_DS.csv"
bf = pd.read_csv(dataset)
bf
```

```python
bf.rename(str.lower, axis="columns", inplace=True)
bf
```

```python
bf.info()
bf
```

```python
columns=["impostos", "beneficios", "vt", "vr", "avaliação do funcionário"]
bf[columns] = bf[columns].apply(lambda x: x.str.replace(",", ".").astype("float"))
bf.info()
```

```python
bf["salario liquido"] = bf["salario base"] - bf["impostos"] + bf["beneficios"] + bf["vt"] + bf["vr"]
bf
```

```python
bf["área"].value_counts(ascending=False)
```

```python
bf.groupby("área")["salario base"].max()
```

```python
bf.groupby("nível")["salario liquido"].agg(["min","max","mean", "sum"])
```
